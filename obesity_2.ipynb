```python
import pandas as pd

# Sample DataFrame
data['NObeyesdad'] = {
    'Category_Column': ['Overweight_Level_II', 'Normal_Weight', 'Insufficient_Weight', 
                        'Obesity_Type_III', 'Obesity_Type_II', 'Overweight_Level_I', 'Obesity_Type_I']
}

df = pd.DataFrame(data['NObeyesdad'])

# Mapping dictionary for categorical to numerical values
mapping_dict = {
    'Overweight_Level_II': 2,
    'Normal_Weight': 0,
    'Insufficient_Weight': 6,
    'Obesity_Type_III': 5,
    'Obesity_Type_II': 4,
    'Overweight_Level_I': 1,
    'Obesity_Type_I': 3
}

# Replace the entire column with numerical values
df['NObeyesdad'] = df['NObeyesdad'].map(mapping_dict)

# Print the modified DataFrame
print(df)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
ohe = OneHotEncoder(sparse_output= False)
ohe
```

```python
df['Gender'] = ohe.fit_transform(df['Gender'].values.reshape(-1,1))
df['family_history_with_overweight'] = ohe.fit_transform(df['family_history_with_overweight'].values.reshape(-1,1))
df['FAVC'] = ohe.fit_transform(df['FAVC'].values.reshape(-1,1))
df['CAEC'] = ohe.fit_transform(df['CAEC'].values.reshape(-1,1))
df['SMOKE'] = ohe.fit_transform(df['SMOKE'].values.reshape(-1,1))
df['SCC'] = ohe.fit_transform(df['SCC'].values.reshape(-1,1))
df['CALC'] = ohe.fit_transform(df['CALC'].values.reshape(-1,1))
df['MTRANS'] = ohe.fit_transform(df['MTRANS'].values.reshape(-1,1))
# df['NObeyesdad'] = ohe.fit_transform(df['NObeyesdad'].values.reshape(-1,1))
```

```python
df
```

```python
x_train = df.iloc[:,:-1]
x_train
```

```python
y_train = df.iloc[:,-1:]
y_train
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_train,y_train,test_size=0.2,random_state= 42)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
ad = AdaBoostClassifier()
ad
```

```python
ad.fit(x_train,y_train)
```

```python
y_pred = ad.predict(x_test)
y_pred
```

```python
accuracy_score(y_test,y_pred)
```
